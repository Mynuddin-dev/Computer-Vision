##### Morphological Transformations are simple operations based on the shape of an image usually performed on a binary image. It takes our input image and a structuring element(kernel) which decides the nature of the operation.

[Erosion and Dilation | Geeksforgeeks](https://www.geeksforgeeks.org/erosion-and-dilation-morphological-transformations-in-opencv-in-cpp/#:~:text=Morphological%20Transformations%20are%20simple%20operations,the%20nature%20of%20the%20operation.)

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('b.jpg', cv2.IMREAD_GRAYSCALE)
_, mask = cv2.threshold(img, 220, 255, cv2.THRESH_BINARY_INV)

kernal = np.ones((5,5), np.uint8)

dilation = cv2.dilate(mask, kernal, iterations=2)
erosion = cv2.erode(mask, kernal, iterations=2)
opening = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernal)
closing = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernal)
mg = cv2.morphologyEx(mask, cv2.MORPH_GRADIENT, kernal)
th = cv2.morphologyEx(mask, cv2.MORPH_TOPHAT, kernal)

titles = ['image', 'Inverse Threshold', 'dilation', 'erosion', 'opening=apply Erosion > Dilation', 'closing =apply Dilation > Erosion', 'Gradient Morphology', 'Tophat']
images = [img, mask, dilation, erosion, opening, closing, mg, th]

plt.figure(figsize=(15,20))
for i in range(8):
    plt.subplot(4, 2, i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
```
