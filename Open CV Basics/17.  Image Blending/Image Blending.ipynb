```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
orange = cv2.imread("orange.jpg")
print(orange.shape)
Orange =cv2.resize(orange,(512,512))
print(Orange.shape)

cv2.imwrite("Orange1.jpg",Orange[:,:])
cv2.imwrite("Orange2.jpg",Orange[:,:,::-1])

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
orange1 = cv2.imread("Orange1.jpg")
orange2 = cv2.imread("Orange2.jpg")

cv2.imshow("Orange1",orange1)
cv2.imshow("Orange2",orange2)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
combine = np.hstack((Orange[:,:256] , Orange[:,256:,::-1]))

cv2.imshow("Orginal Orange",orange)
cv2.imshow("Orginal combine Orange",combine)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

# generate Gaussian pyramid for orange1
orange1_copy = orange1.copy()
gp_orange1 = [orange1_copy]
for i in range(6):
    orange1_copy = cv2.pyrDown(orange1_copy)
    gp_orange1.append(orange1_copy)


# generate Gaussian pyramid for orange2
orange2_copy = orange2.copy()
gp_orange2 = [orange2_copy]
for i in range(6):
    orange2_copy = cv2.pyrDown(orange2_copy)
    gp_orange2.append(orange2_copy)
    

# generate Laplacian Pyramid for orange1
orange1_copy = gp_orange1[5]
lp_orange1 = [orange1_copy]
for i in range(5, 0, -1):
    gaussian_expanded = cv2.pyrUp(gp_orange1[i])
    laplacian = cv2.subtract(gp_orange1[i-1], gaussian_expanded)
    lp_orange1.append(laplacian)

    
# generate Laplacian Pyramid for orange2
orange2_copy = gp_orange2[5]
lp_orange2 = [orange2_copy]
for i in range(5, 0, -1):
    gaussian_expanded = cv2.pyrUp(gp_orange2[i])
    laplacian = cv2.subtract(gp_orange2[i-1], gaussian_expanded)
    lp_orange2.append(laplacian)

    
# Now add left and right halves of images in each level
orange_pyramid = []
n = 0
for orange1_lap, orange2_lap in zip(lp_orange1, lp_orange2):
    n += 1
    cols, rows, ch = orange1_lap.shape
    laplacian = np.hstack((orange1_lap[:, 0:int(cols/2)], orange2_lap[:, int(cols/2):]))
    orange_pyramid.append(laplacian)
# now reconstruct
orange_reconstruct = orange_pyramid[0]
for i in range(1, 6):
    orange_reconstruct = cv2.pyrUp(orange_reconstruct)
    orange_reconstruct = cv2.add(orange_pyramid[i], orange_reconstruct)
    
cv2.imshow("Orange1" , orange1)
cv2.imshow("Orange2" , orange2)
cv2.imshow("Combine" , combine)
cv2.imshow("Reconstruct" , orange_reconstruct)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
